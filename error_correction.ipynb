```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
%matplotlib inline
```

```python
IMG_ROOT = 'D:\Codes\mask_recognition\data/train'
data=pd.read_csv(r'D:\Codes\mask_recognition\xml_df')
data
```

```python
# Get image as numpy array
def load_image(name, path):
    img_path = path +'/'+ name + '.jpg'
    # img = cv2.imread(img_path)
    img = mpimg.imread(img_path)
    return img

# Plot numpy array
def plot_image(img):
    plt.imshow(img)
    
    plt.title(img.shape)
    
# Plot a grid of examples
def plot_grid(img_names, img_root, rows=5, cols=5):
    fig = plt.figure(figsize=(25,25))
    
    for i,name in enumerate(img_names):
        fig.add_subplot(rows,cols,i+1)
        img = load_image(name, img_root)
        plot_image(img)
        
    plt.show()
```

```python
plot_grid(data['filename'][:25], IMG_ROOT)
```
